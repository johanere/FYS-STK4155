```python
import pandas as pd
import sklearn.preprocessing as sklpre
import sklearn.model_selection as sklms
import sklearn.metrics as sklmet
import matplotlib.pyplot as plt
import numpy as np
import keras
```

```python
df = pd.read_csv("NN_gridsearch_results/NN_gridsearch_RS0_dropCTrue__f1macro1012_0014.csv", sep=",")
```

```python
hyper_measure=df[["param_act","param_batch_size","param_dropout","param_beta_1","param_beta_2","param_decay","param_dropout","param_epochs","param_lr","param_nl1","param_nl2","param_nl3","param_nn1","param_nn2","param_nn3","param_l1","param_l2","mean_test_score"]]
```

```python
hyper_measure
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10 * 1.618, 10))
sns.heatmap(hyper_measure.corr(), cmap="YlGnBu", ax=ax, square=True, annot=True, cbar=False)
fig.tight_layout()
plt.title("Corrolation matrix")
plt.show()
```

```python
hp=hyper_measure.drop(["param_nl1","param_nl2","param_nl3","param_nn1","param_nn2","param_nn3","param_l1","param_l2"],axis=1)
```

```python
initial_parameters=hp.drop(["param_dropout","param_beta_1","param_beta_2","param_decay","param_dropout"],axis=1)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})

fig, ax = plt.subplots(figsize=(10 * 1.618, 10))
sns.heatmap(initial_parameters.corr(), cmap="YlGnBu", ax=ax, square=True, annot=True, cbar=False)
ax.set_ylim([4, 0])
ax.set_xlim([0, 4])
ax.set_xticklabels(['Batch size',"Epochs","Learning Rate","F1 score"])
ax.set_yticklabels(['Batch size',"Epochs","Learning Rate","F1 score"])
plt.title("Corrolation matrix of initial hyper-parameter grid-search")
plt.savefig("../Results/inital_hyperparam.pdf")
plt.show()
```
